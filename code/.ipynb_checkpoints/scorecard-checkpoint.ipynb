```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import scorecardpy as sc
from sklearn.model_selection import train_test_split
import statsmodels.api as sm
from sklearn.metrics import roc_auc_score, roc_curve
```

```python
train =pd.read_csv("../data/train_woe.csv")
test = pd.read_csv("../data/test_woe.csv")
```

```python
train_X = train.drop("target", axis=1)
train_X = sm.add_constant(train_X)
train_y = train["target"]

test_X = test.drop("target", axis=1)
test_X = sm.add_constant(test_X)
```

```python
#逻辑回归
log_model = sm.Logit(train_y, train_X)
log_result = log_model.fit()
log_result.summary()
```

```python
#比对预测值
test["predicted_target"] = log_result.predict(test_X)
test[["predicted_target","target"]].head(10)
```

```python
#模型评估
auc = roc_auc_score(test["target"], test["predicted_target"])
fpr, tpr, threshold = roc_curve(test["target"], test["predicted_target"])
ks = max(tpr - fpr)
auc,ks
```

0.4<=KS<=0.7，说明模型的总体表现良好，有一定区分度。AUC也>0.5，但不算特别高。
